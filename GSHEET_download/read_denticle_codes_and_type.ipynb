```python
nested = {}
for name1, group1 in types.groupby('Associated Gross Type', dropna=False):
        if pd.isna(name1):
            for elem in list(group1['dropdown menu']):
                if pd.isna(elem): elem = 'NA'
                nested[elem] = None
        else:
            nested[name1] = {}
            for elem in list(group1['dropdown menu']):
                 nested[name1][elem] = None
nested
```
